Q3 Explain the following term :
* The difference in np.array, np.asarray and np.asanyarray.
* The difference between Deep copy and shallow copy.

I'd be glad to explain the differences between `np.array`, `np.asarray`, and `np.asanyarray`, as well as deep copy and shallow copy.

**1. `np.array`, `np.asarray`, and `np.asanyarray`**

All three functions are used to create NumPy arrays, but they have subtle differences in their behavior:

- **`np.array(object)`:** This is the most general function for creating arrays. It takes any object that can be converted to an array and returns a new NumPy array. If the input is already a NumPy array, it returns a copy.
- **`np.asarray(object)`:** Similar to `np.array`, but it tries to convert the input to a NumPy array without copying the data if possible. If the input is already a NumPy array, it returns a reference to the same array.
- **`np.asanyarray(object, dtype=None, order=None)`:** This function is more flexible than `np.asarray` because it allows you to specify the desired data type (`dtype`) and memory layout (`order`) of the output array. If the input is already a NumPy array with the specified dtype and order, it returns a reference to the same array. Otherwise, it creates a new array.

**2. Deep Copy and Shallow Copy**

When copying NumPy arrays, you can choose between deep copy and shallow copy. The difference lies in how the data is copied:

- **Shallow Copy:** A shallow copy creates a new array object, but the data is shared between the original and the copy. If you modify the data in one array, it will also affect the other array.
- **Deep Copy:** A deep copy creates a new array object and copies all the data elements. Modifying the data in one array will not affect the other array.

Here's a table summarizing the key differences:

| Function | Behavior | Data Copy |
|---|---|---|
| `np.array` | Creates a new array | Always deep copy |
| `np.asarray` | Tries to convert without copying | Shallow copy if possible, deep copy otherwise |
| `np.asanyarray` | Similar to `np.asarray` | Shallow copy if possible, deep copy otherwise |
| Deep Copy | Creates a new array with independent data | Deep copy |
| Shallow Copy | Creates a new array with shared data | Shallow copy |

**Choosing the Right Function**

- **`np.array`:** Use this function when you want to ensure that the output is always a new array, regardless of the input type.
- **`np.asarray`:** Use this function when you want to avoid unnecessary copying if the input is already a NumPy array.
- **`np.asanyarray`:** Use this function when you need to control the data type or memory layout of the output array.

**Deep Copy vs. Shallow Copy**

- Use deep copy when you want to modify the data in one array without affecting the other.
- Use shallow copy when you want to share data between multiple arrays and avoid unnecessary copying.

By understanding these differences, you can choose the appropriate function and copy method for your specific use case.

14. Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and 
Perform the following tasks using Matplotlib:

a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and 
right y-axis for 'Humidity').

b) Label the x-axis as 'Date'.

c) Set the title of the plot as 'Temperature and Humidity Over Time.'

```python
import pandas as pd
import matplotlib.pyplot as plt

dates = pd.date_range(start='2023-01-01', periods=30)
temperature = np.random.randint(60, 90, size=30)
humidity = np.random.randint(30, 80, size=30)

df = pd.DataFrame({'Date': dates, 'Temperature': temperature, 'Humidity': humidity})

fig, ax1 = plt.subplots()

ax1.plot(df['Date'], df['Temperature'], color='blue', label='Temperature')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (F)', color='blue')

ax2 = ax1.twinx()
ax2.plot(df['Date'], df['Humidity'], color='red', label='Humidity')
ax2.set_ylabel('Humidity (%)', color='red')

plt.title('Temperature and Humidity Over Time')

plt.legend()

plt.tight_layout()
plt.show()
```

Q17 Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randn(100)
y = np.random.randn(100)

quadrant = np.where(x >= 0, np.where(y >= 0, 1, 4), np.where(y >= 0, 2, 3))

sns.scatterplot(x=x, y=y, hue=quadrant, palette='bright')

plt.legend(title='Quadrant')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.title('Quadrant-wise Scatter Plot')

plt.show()
```

Q18.With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'.

Q19. Using Bo+eh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical 
Bar Chart'8

```python
import numpy as np
import bokeh
from bokeh.plotting import figure, output_file, show

x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

p = figure(title="Sine Wave Function", x_axis_label='X', y_axis_label='Y')

p.line(x, y, line_width=2)

p.grid.grid_line_alpha = 0.5

output_file("sine_wave.html")
show(p)
```
